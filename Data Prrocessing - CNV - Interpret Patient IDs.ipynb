```python
import pandas as pd

datapath ="C:/Users/garim/Documents/Capstone/Full data/Raw/cnv.window_1M.csv"
cnv = pd.read_csv(datapath)
```

```python
tss_map = sample_map = pd.read_excel("C:/Users/garim/Documents/Capstone/Full data/tss_map.xlsx",sheet_name='tss_map',header=0,converters={'tssC':str,'TissueSourceSite':str},na_values = [], keep_default_na  = False)

sample_map = pd.read_excel("C:/Users/garim/Documents/Capstone/Full data/sample_map.xlsx",sheet_name='sample_map',header=0,converters={'sampleC':str,'sample':str},na_values = [], keep_default_na  = False)
```

```python
tss_map['tssC'] = tss_map['tssC'].astype(str)
sample_map['sampleC']=sample_map['sampleC'].astype(str)
```

```python
tcnv = cnv.transpose()
header = tcnv.iloc[0]
tcnv = tcnv[1:]
tcnv.columns = header
```

```python
tcnv.shape
```

Splitting Patient ID into components to obtain Disease and sample information

```python
tcnv[['study','tssC','participant','sample_vialC','portion_analyte','plate','center']] = pd.DataFrame(tcnv.index.str.split('.').values.tolist(), index= tcnv.index)
tcnv.iloc[0:5,:]
```

```python
tcnv['sampleC'] =  tcnv['sample_vialC'].str.slice(0,2)
tcnv['vial']= tcnv['sample_vialC'].str.slice(2,3)

tcnv.iloc[0:5,:]
```

Pulling the meaning of TSS and Sample codes under 'tss' and 'sample' columns respectively

```python
tcnv = tcnv.reset_index()
tcnv['tssC'] = tss_map['tssC'].astype(str)
tcnv = pd.merge(tcnv, tss_map, how = 'left', on = 'tssC')
tcnv = pd.merge(tcnv, sample_map, how = 'left', on = 'sampleC')
```

```python
tcnv=tcnv.drop(['sample_vialC'], axis = 1)
```

Exporting the modified TCGA CNV dataframe to TCNV_TCGA csv file

```python
tcnv.iloc[:5,:]
```
